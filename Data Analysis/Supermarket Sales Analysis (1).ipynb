```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
sns.barplot(x="Branch", y="Total" , data= data ,hue = "Branch")
plt.title("Distrbution of Sales Across Branches")
plt.xlabel("Branches")
plt.ylabel("Total")
plt.show()
```

```python
sns.barplot(x="City" , y="Total" , data = data , hue = 'City')
plt.title("Sales Revenue by City")
plt.show()
```

```python
sns.barplot(y="Customer type", x="Payment" , data = data, hue ="Payment")
```

```python
plt.figure(figsize=(8,8))
sns.countplot(x='Product line', hue='Gender', data=data ,palette='Set1')
plt.xticks(rotation=45)  
plt.show()
```

```python
sns.barplot(x="Product line" , y = "Total" , data= data , hue="Product line")
plt.xticks(rotation=45);
```

```python
data['Time'] = pd.to_datetime(data['Time'])

data['Hour'] = data['Time'].dt.hour
sns.countplot(x="Hour", hue="Payment" , data = data)
```

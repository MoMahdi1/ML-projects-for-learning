```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_csv("mpg-unclean.csv")
cars.head()
```

```python
cars.info()
```

```python
cars.describe().round(2)
```

```python
cars.plot(subplots = True , figsize=(20,10))
```

```python
cars.columns
```

```python
cars.rename(columns={'model year':'model_year'} , inplace = True)
```

```python
cars.head()
```

```python
cars["origin"].value_counts()
```

```python
cars.origin.replace({"United States" :"usa"} , inplace = True)
```

```python
cars.head()
```

```python
cars["horsepower"]=cars["horsepower"].str.replace(" hp","")
```

```python
cars.head()
```

```python
cars.horsepower.replace("Not available", np.nan , inplace= True)
```

```python
cars["horsepower"] = cars["horsepower"].astype("float")
```

```python
cars.dtypes
```

```python
cars.name
```

```python
cars.name = cars.name.str.lower().str.strip()
```

```python
cars.head()
```

```python
cars.model_year.replace(1973,73 , inplace = True)
```

```python
cars.weight.max()
```

```python
cars.loc[cars.weight.idxmax() , "weight"] =2300
```

```python
cars.describe()
```

```python
cars.mpg.min()
```

```python
cars.loc[cars['mpg'].idxmin() , "mpg"] = 1 / cars.loc[cars['mpg'].idxmin() , "mpg"]
```

```python
cars.describe().round(2)
```

```python
cars.isna().sum()
```

```python
cars.dropna(subset=["horsepower"], inplace = True)
```

```python
cars.isna().sum()
```

```python
cars.duplicated().sum(axis=0)
```

```python
cars.drop_duplicates(inplace=True)
```

```python
cars.duplicated().sum(axis=0)
```

```python
cars.to_csv("data/mpg-cleaned.csv")
```
